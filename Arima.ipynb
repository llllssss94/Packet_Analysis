```python
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
data = pd.read_excel('./data/export.xlsx')
```

```python
data = data[data.length >= 200]
```

```python
data = data[['length', 'label', 'timestamp']].groupby(['timestamp'], as_index=False).mean()
```

```python
data.label = (data.label > 0.5)*1
```

```python
sns.scatterplot(x=range(len(data.length)), y=data.length)
```

```python
data_normal = data[data.label == 0]
```

```python
sns.scatterplot(x=range(len(data_normal.length)), y=data_normal.length)
```

```python
data_mal = data[data.label == 1]
sns.scatterplot(x=range(len(data_mal.length)), y=data_mal.length)
```
